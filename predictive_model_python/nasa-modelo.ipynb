# Preparing train and test data

```python

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
x = x.reset_index()
```

```python
x.iloc[:,0:1251] = clean_dataset(x.iloc[:,0:1251])
```

```python
x.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
#Replacing Na
x.iloc[:,0:1251].fillna(0) 

#Separando teste e treino
y_train = x['class'].sample(frac=0.6, replace=True, random_state=1)
index_train = y_train.index.values.tolist()
index = x.index.values.tolist()
index_test =  [test for test in index if test not in index_train]
y_test = x['class'][index_test]


x_train = x.iloc[index_train, 0:1251]
x_test = x.iloc[index_test, 0:1251]
```

```python
x.iloc[:,0:1251].fillna(0) 
```

```python
x_train= np.nan_to_num(x_train)
y_train= np.nan_to_num(y_train)
x_test= np.nan_to_num(x_test)
y_test= np.nan_to_num(y_test)
```
